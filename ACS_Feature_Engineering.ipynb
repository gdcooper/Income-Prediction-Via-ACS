However, another nonprofit census reporting organization, Census Reporter, which analyzes data from the ACS, used the income tables for Loving County to report a median income of $44,821. So, I'll use that figure as an estimate for Loving's missing data.

I want to create 5 categories for Median Income: Poverty, Low, Medium, High, and Very High. The median is 60,458, and the official boundary for poverty in the U.S. for a family of 3 is 25,820, according to the Office of the Assistant Secretary for Planning and Education, so that will be the boundary for a 'Poverty' Category. Everything between that and the 25th percentile at 51,814 will be 'Low', from there to the 75th percentile at 70,376 will be 'Medium', up to 100,000 will be 'High', and anything above that will be 'Very High'.

```python
# Creating income categories
bins = [0, 25820, 51814, 70376, 100000, float("inf")]
labels = ["Poverty", "Low", "Medium", "High", "Very High"]
# Add new category column to acs_raw data
acs_raw['Income Category'] = pd.cut(acs_raw["Median Income"], bins=bins, labels=labels)
```

```python
# Display results
acs_raw.head()
```

```python
# Convert to category
acs_raw['Income Category'] = acs_raw['Income Category'].astype('category')
# Check type
acs_raw['Income Category'].dtype
```

```python
# Should be a good distribution for models to work with
acs_raw['Income Category'].value_counts()
```

```python
acs_raw['Population Below Poverty Line'].describe()
```

```python
# New variable: High vs. Low poverty rate
acs_raw['Poverty Rate (%)'] = (acs_raw['Population Below Poverty Line'] / acs_raw['Total Population']) * 100
```

The USDA Economic Research Service describes a high area poverty rate as 20% or higher, so I'll make a binary variable of 'High Poverty rate' vs. 'Low Poverty rate', so it'll essentially be a variable saying, 'Is this area's poverty rate high? Yes or No, 0 or 1'.

```python
acs_raw['Poverty Rate Classification'] = np.where(acs_raw['Poverty Rate (%)'] > 20, 'High Poverty', 'Low Poverty')
```

```python
acs_raw['Poverty Rate Classification'].head()
```

```python
# Check results
acs_raw['Poverty Rate Classification'].value_counts()
```

```python
# Create new column based on degree sum
acs_raw['Total Degrees'] = acs_raw[["Bachelor's Degree", "Master's Degree", 'Professional School Degree', 'Doctorate']].sum(axis=1)
```

```python
# Create percentage verusus total population
acs_raw['Degree Percentage (%)'] = (acs_raw['Total Degrees'] / acs_raw['Total Population']) * 100
```

```python
sns.boxplot(data = acs_raw, x = "Degree Percentage (%)")
```

```python
acs_raw['Degree Percentage (%)'].describe()
```

```python
# Create category for Low, Medium, High Degree Rates
# According to census Bureau, U.S. total is 32% for people over 25, as of 2023
# Going to use 0 to 15 as 'Low', 15 to 25 as 'Medium', and anything above 25 as 'High'
bins2 = [0, 15, 25, float("inf")]
labels2 = ["Low Degree Attainment", "Medium Degree Attainment", "High Degree Attainment"]
# Add new category column to acs_raw data
acs_raw['Degree Attainment'] = pd.cut(acs_raw["Degree Percentage (%)"], bins=bins2, labels=labels2)
```

```python
# Check data
acs_raw['Degree Attainment'].value_counts()
```

```python
# Next category will be low, medium, high People of Color (POC) population
# Plus percentages for each individual race as recorded by the ACS
# This is important to consider because compared to average White Americans, virtually every other group is paid less and more likely to be in poverty (Pew Research Center)
#
acs_raw['POC Population (%)'] = (acs_raw[['Black/African American', 'Asian', 'American Indian/Alaskan', 'Some Other Race', 'Two or More Races']].sum(axis=1) / acs_raw['Total Population']) * 100
acs_raw['White Population (%)'] = (acs_raw['White'] / acs_raw['Total Population']) * 100
acs_raw['Black/African American Population (%)'] = (acs_raw['Black/African American'] / acs_raw['Total Population']) * 100
acs_raw['Asian Population (%)'] = (acs_raw['Asian'] / acs_raw['Total Population']) * 100
acs_raw['American Indian/Alaskan Population (%)'] = (acs_raw['American Indian/Alaskan'] / acs_raw['Total Population']) * 100
acs_raw['Hawaiian/Pacific Islander Population (%)'] = (acs_raw['Hawaiian/Pacific Islander'] / acs_raw['Total Population']) * 100
acs_raw['Some Other Race Population (%)'] = (acs_raw['Some Other Race'] / acs_raw['Total Population']) * 100
acs_raw['Two or More Races Population (%)'] = (acs_raw['Two or More Races'] / acs_raw['Total Population']) * 100
```

```python
acs_raw['POC Population (%)'].describe()
```

```python
bins3 = [0, 10, 20, float("inf")]
labels3 = ["Low", "Medium", "High"]
# Add new category column to acs_raw data
acs_raw['POC Population Category'] = pd.cut(acs_raw["POC Population (%)"], bins=bins3, labels=labels3)
```
